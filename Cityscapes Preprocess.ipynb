```python
def label_categorization(img, num_classes, img_size):
  primary_mask = torch.zeros((num_classes, 1024, 2048))
  channled_mask = torch.zeros((num_classes, img_size, img_size)) 
  
  temp_transform = transforms.Compose([transforms.Resize((512, 512)),])
  
  for i in range(num_classes):
    primary_mask[i][img.squeeze() == i] = 1              # Converting [1, 1024, 2048] to [num_classes, 1024, 2048]
    channled_mask[i] = temp_transform(torch.unsqueeze(primary_mask[i], dim = 0))   # resizing [num_classes, 1024, 2048] to [num_classes, img_size = 512, img_size = 512]
    channled_mask[i] = (channled_mask[i] > 0.50).float() # thresholding to have just 1 and 0 in each channel
  
  return channled_mask
```
